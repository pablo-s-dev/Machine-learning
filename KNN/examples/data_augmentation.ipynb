```python
def data_augmentation(img, amount):

    img = np.expand_dims(img, axis=0)
    
    datagen = ImageDataGenerator(
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        
        fill_mode='nearest'
    )
    datagen.fit(img)
    i = 0
    imgs = []
    for x_batch in datagen.flow(img, batch_size=amount):
        imgs.append(x_batch)
        i += 1
        if i == amount:
            break

    return np.array(imgs)
```

```python
for i, img in enumerate(train_img):

    label = train_labels[i].replace(".pet", "_pet")

    dir_name = f"D:\Computer Science\TCC Ciência e Tecnologia\Data\Perdidos_square_cut_augmented\{label}"

    if not os.path.exists(dir_name):
        os.makedirs(dir_name)

    new_imgs = data_augmentation(img/255, 10)

    for j, img in enumerate(new_imgs):

        filename = train_filenames[i].replace(".jpg", "")

        plt.imsave(f"{dir_name}\\{filename}_aug{j+1}.jpg", img[0])  
```
